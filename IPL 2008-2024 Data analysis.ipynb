```python
import pandas as pd

# Correct file paths to the CSV files
matches_path = r'C:\Users\srava\OneDrive\Desktop\IPL 2008-2024\IPL Dataset\matches.csv'
deliveries_path = r'C:\Users\srava\OneDrive\Desktop\IPL 2008-2024\IPL Dataset\deliveries.csv'

# Read the CSV files using read_csv
matches = pd.read_csv(matches_path)
deliveries = pd.read_csv(deliveries_path)

# Display the first few rows of each dataset
print("Matches Data:")
print(matches.head())

print("\nDeliveries Data:")
print(deliveries.head())
```

```python
print("Matches shape:", matches.shape)
print("Deliveries shape:", deliveries.shape)

print("\nMatches columns:\n", matches.columns)
print("\nDeliveries columns:\n", deliveries.columns)
```

```python
print("\nMissing values in Matches:\n", matches.isnull().sum())
print("\nMissing values in Deliveries:\n", deliveries.isnull().sum())
```

```python
print("Total seasons played:", matches['season'].nunique())
print("Seasons:", matches['season'].unique())
```

```python
import matplotlib.pyplot as plt

season_counts = matches['season'].value_counts().sort_index()
print(season_counts)

# Plotting
plt.figure(figsize=(10, 4))
season_counts.plot(kind='bar', color='skyblue')
plt.title("Matches Played Per Season")
plt.xlabel("Season")
plt.ylabel("Number of Matches")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
team_wins = matches['winner'].value_counts()
print(team_wins)

# Plotting
plt.figure(figsize=(10, 5))
team_wins.plot(kind='bar', color='lightgreen')
plt.title("Most Matches Won by Teams")
plt.xlabel("Team")
plt.ylabel("Wins")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
top_batsmen = deliveries.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False).head(10)
print(top_batsmen)

# Plot
top_batsmen.plot(kind='bar', color='orange', figsize=(10, 5))
plt.title("Top 10 Run Scorers in IPL (2008–2024)")
plt.xlabel("Batsman")
plt.ylabel("Total Runs")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Filter rows where wicket actually fell
wickets = deliveries[deliveries['is_wicket'] == 1]

# Remove run outs and invalid dismissals for bowling analysis
valid_dismissals = ['bowled', 'caught', 'lbw', 'stumped', 'caught and bowled', 'hit wicket']
wickets = wickets[wickets['dismissal_kind'].isin(valid_dismissals)]

top_bowlers = wickets['bowler'].value_counts().head(10)
print(top_bowlers)

# Plot
top_bowlers.plot(kind='bar', color='purple', figsize=(10, 5))
plt.title("Top 10 Wicket Takers in IPL (2008–2024)")
plt.xlabel("Bowler")
plt.ylabel("Wickets")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Filter players with at least 500 balls faced
balls_faced = deliveries.groupby('batter')['ball'].count()
total_runs = deliveries.groupby('batter')['batsman_runs'].sum()

strike_rate = (total_runs / balls_faced) * 100
qualified_players = strike_rate[balls_faced >= 500].sort_values(ascending=False).head(10)

print(qualified_players)

# Plot
qualified_players.plot(kind='bar', color='teal', figsize=(10, 5))
plt.title("Top 10 Strike Rates (Min 500 balls)")
plt.xlabel("Batsman")
plt.ylabel("Strike Rate")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
